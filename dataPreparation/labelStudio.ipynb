## Create json with image path and traditional ocr outputs for label studio

```python
import csv
import json
from pathlib import Path

folders = ["test/rawCropTest", "train/rawCropTrain", "val/rawCropVal"]
csvFiles = ["data/OCRcrops/test/tradOutputTest/ocrResultsTest.csv", "data/OCRcrops/train/tradOutputTrain/ocrResultsTrain.csv", "data/OCRcrops/val/tradOutputVal/ocrResultsVal.csv"]

for folder, csvs in zip(folders, csvFiles):

    # Root must match LABEL_STUDIO_LOCAL_FILES_DOCUMENT_ROOT
    root = Path("/home/daniel/Documents/deepLearning/Project2_TrOCR/data/OCRcrops")

    # current folder
    image_folder = root / folder

    csv_file = csvs
    output_json = f"{Path(folder).name}LS.json"

    data = []

    with open(csv_file) as f:
        reader = csv.DictReader(f)
        for row in reader:
            img_path = image_folder / row["filename"]

            # Convert to relative path from root
            relative_path = img_path.relative_to(root)

            data.append({
                "data": {
                    "image": f"/data/local-files/?d={relative_path}",
                    "text": row["best_text"] if row["best_text"] != "null" else ""
                }
            })

    with open(output_json, "w") as f:
        json.dump(data, f, indent=2)
```
